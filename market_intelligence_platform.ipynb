## Motor de Procesamiento y Análisis Semántico a Gran Escala con Arquitectura RAG

Pipeline de Big Data y Motor Generativo Asimétrico para la Auditoría Automatizada de la "Voz del Cliente" (Voice of Customer - VoC).

---

## Descripción General

Este proyecto consiste en el diseño, desarrollo e implementación de una plataforma de datos de extremo a extremo (End-to-End) que unifica la ingeniería de datos a gran escala (Big Data) con la Inteligencia Artificial Generativa y Procesamiento de Lenguaje Natural (NLP).

El sistema ingesta de forma masiva millones de registros no estructurados de retroalimentación de clientes (reseñas, quejas y opiniones), limpia y procesa la información mediante flujos analíticos, y extrae metadatos críticos de manera automatizada.

Finalmente, expone una consola inteligente basada en una arquitectura RAG (Retrieval-Augmented Generation) que permite a los directores de negocio e investigadores de mercado interrogar analíticamente el repositorio mediante lenguaje natural, obteniendo reportes consolidados y libres de alucinaciones.

---

## Problemática / Necesidad de Negocio

Las corporaciones globales y grandes cadenas de Retail e E-commerce se enfrentan al reto de procesar diariamente volúmenes masivos de datos textuales provenientes de sus canales de atención y venta. Analizar manualmente estos textos para extraer insights accionables es operativamente inviable y costoso.

Las soluciones tradicionales de Business Intelligence (BI) fallan en este entorno porque solo pueden medir métricas estructuradas cuantitativas (como calificaciones de estrellas), perdiendo por completo el contexto cualitativo (el porqué de la insatisfacción, fallas repetitivas en productos específicos o alertas tempranas de fraude). Por otro lado, implementar Inteligencia Artificial Generativa directamente sobre estos hilos de datos sin control provoca altos costos de procesamiento en la nube, riesgos de filtración de datos y respuestas inexactas (alucinaciones del modelo).

---

## Solución Propuesta (Arquitectura de Alta Fidelidad)

Para solucionar este cuello de botella corporativo, se diseñó una arquitectura híbrida de alto rendimiento:

* **Fase de Big Data (Databricks + PySpark):**
    Ingesta masiva del dataset original no estructurado para realizar tareas de normalización, limpieza y transformación mediante procesamiento distribuido en memoria, garantizando la escalabilidad empresarial.

* **Fase de Almacenamiento Relacional y Semántico (PostgreSQL + pgvector):** 
    Migración de los datos procesados a una base de datos robusta, donde la información estructurada se modela de forma relacional y las opiniones textuales se transforman en embeddings vectoriales, almacenándose en la misma base mediante la extensión pgvector para unificar consultas analíticas SQL y búsquedas semánticas.

* **Fase de Inteligencia y Orquestación (Python, LangChain & LLM):**
    Un microservicio interno construido en Python que fragmenta la información, genera vectores y orquesta las llamadas hacia un LLM Open Source de alta velocidad (vía Groq/Ollama), sirviendo como el motor RAG.

* **Fase de Negocio y Entrega Web (PHP + React.js):** 
    Un backend robusto en PHP que gobierna la aplicación, gestiona bitácoras de auditoría de preguntas y consume la API de IA, exponiendo los datos hacia un dashboard interactivo en React.js que combina analítica tradicional y una terminal de IA conversacional.

```python
import os
import sys
import time
from getpass import getpass

import zipfile
import json

import re
import tqdm

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from pyspark.sql import SparkSession
from pyspark.sql.functions import col, when, length, regexp_replace, split, size, round

from langchain_text_splitters import RecursiveCharacterTextSplitter

import chromadb
from chromadb.utils import embedding_functions

from groq import Groq
```

```python
# Definir rutas de los archivos
zip_path = "electronics_5.zip"
output_sample_path = "electronics_sample_30k.json"

print("Entorno configurado correctamente.")
```

```python
with zipfile.ZipFile(zip_path, 'r') as z:
    # Listar todos los archivos dentro del zip
    file_list = z.namelist()
    print("Archivos encontrados dentro del ZIP:")
    for file in file_list:
        print(f"- {file}")
        
# Guardamos el nombre del primer archivo JSON encontrado
json_file_name = [f for f in file_list if f.endswith('.json')][0]
```

```python
# Definir el tamaño de la muestra deseada
sample_size = 30000
sampled_lines = []
rng = np.random.default_rng(seed=42)

print("Iniciando lectura por streaming desde el ZIP...")

with zipfile.ZipFile(zip_path, 'r') as z:
    with z.open(json_file_name) as f:
        # Usamos un contador para aplicar un muestreo probabilístico simple o Reservoir Sampling
        for i, line in enumerate(f):
            # En las primeras 30,000 líneas, llenamos la lista directamente
            if i < sample_size:
                sampled_lines.append(json.loads(line.decode('utf-8')))
            else:
                # Para las líneas siguientes, decidimos probabilísticamente si reemplaza a una existente
                # Esto asegura que la muestra sea verdaderamente representativa de todo el archivo
                r = rng.integers(0, i + 1)
                if r < sample_size:
                    sampled_lines[r] = json.loads(line.decode('utf-8'))
            
            # Barra de progreso simulada cada 250,000 líneas
            if i % 250000 == 0 and i > 0:
                print(f"Procesadas {i} líneas del archivo original...")

print(f"\n¡Mapeo completado! Se han extraído {len(sampled_lines)} registros aleatorios.")

# Convertir la lista de diccionarios a un DataFrame de Pandas para verificar
df_sample = pd.DataFrame(sampled_lines)
print(f"Estructura del DataFrame local: {df_sample.shape}")
```

```python
# Guardar el DataFrame en formato JSON Lines para mantener compatibilidad
df_sample.to_json(output_sample_path, orient='records', lines=True)
print(f"Muestra guardada con éxito en: {output_sample_path}")
```

#### Tipos de Datos (Dtypes) e Ingeniería de Atributos (*Feature Engineering*)

* **Normalización Estructural de Utilidad (`helpful`):**
    * *Estado Crítico Inicial:*
        El atributo original viene serializado como un objeto complejo o arreglo de Python (ej. `[3, 3]`, `[1, 1]`). Almacenar estructuras compuestas (listas/arrays) dentro de un motor relacional como PostgreSQL sin un propósito de matriz penaliza el rendimiento e incumple directamente la **Primera Forma Normal (1FN)** de las bases de datos relacionales.
    * *Resolución de Ingeniería:*
        Se aplica una transformación lambda vectorizada mediante Pandas para descomponer analíticamente el array en tres campos planos e independientes de alta eficiencia computacional:
        1.  `helpful_votes` (`int64`): Volumen de usuarios que validaron positivamente la utilidad de la opinión.
        2.  `total_votes` (`int64`): Universo total de interacciones/votos registrados en la reseña.
        3.  `helpful_ratio` (`float64`): Métrica derivada calculada mediante la relación $\frac{\text{helpful\_votes}}{\text{total\_votes}}$, protegiendo el flujo contra errores fatales de división por cero (`NaN`) mediante condiciones lógicas controladas.
* **Redundancia Temporal (`reviewTime` vs `unixReviewTime`):**
    El campo `reviewTime` se presenta como un string de texto humano (ej. *"06 8, 2007"*). Procesar cadenas de texto para análisis cronológicos en SQL consume ciclos de cómputo masivos de forma innecesaria. 
    * *Resolución de Ingeniería:*
        Se determina conservar únicamente `unixReviewTime` como un entero puro (`int64`), delegando la conversión a un tipo de dato nativo `TIMESTAMP` o `DATE` (YYYY-MM-DD) al momento de la inserción en la base de datos PostgreSQL, optimizando la indexación temporal.

#### Limpieza de Texto y Purga de Nulos

```python
## Función de limpieza de texto para NLP
def clean_text(text):
    if not isinstance(text, str):
        return ""
    # Remover etiquetas HTML remanentes (muy común en web scraping)
    text = re.sub(r'<.*?>', '', text)
    # Normalizar espacios en blanco, tabulaciones y saltos de línea a un solo espacio
    text = re.sub(r'\s+', ' ', text)
    return text.strip()
```

```python
# Leer el archivo CRUDO para simular la transformación desde el estado original al estado limpio dentro de Spark
path_json_raw = "electronics_sample_30k.json" 
df_raw = spark.read.json(path_json_raw)

# Pipeline de Transformación Nativo de Spark (Simulación Databricks)
## 1. Renombrar columnas a snake_case
## 2. Desempaquetar el array 'helpful' en columnas numéricas independientes
## 3. Calcular la métrica derivada 'helpful_ratio'
## 4. Calcular el conteo de palabras aproximado
## 5. Limpiar el texto de 'summary' y 'review_text' usando expresiones regulares para eliminar etiquetas HTML
## 6. Filtrar reseñas sin texto (word_count = 0) para evitar registros vacíos y errores en el análisis posterior
## 7. Eliminar las columnas originales 'helpful', 'summary' y 'review_text' para mantener solo las columnas limpias y normalizadas
df_spark_clean = df_raw \
    .withColumnRenamed("reviewerID", "reviewer_id") \
    .withColumnRenamed("reviewerName", "reviewer_name") \
    .withColumnRenamed("reviewText", "review_text") \
    .withColumnRenamed("unixReviewTime", "unix_review_time") \
    .withColumnRenamed("reviewTime", "review_time") \
    .withColumn("helpful_votes", col("helpful").getItem(0)) \
    .withColumn("total_votes", col("helpful").getItem(1)) \
    .withColumn("helpful_ratio", 
                when(col("total_votes") > 0, round(col("helpful_votes") / col("total_votes"), 4))
                .otherwise(0.0)) \
    .withColumn("word_count", 
                when(col("review_text").isNotNull(), size(split(col("review_text"), " ")))
                .otherwise(0)) \
    .filter(col("word_count") > 0) \
    .withColumn("clean_summary", regexp_replace(col("summary"), "<.*?>", "")) \
    .withColumn("clean_summary", 
                when((col("clean_summary") == "") | col("clean_summary").isNull(), "No Title")
                .otherwise(col("clean_summary"))) \
    .withColumn("clean_review_text", regexp_replace(col("review_text"), "<.*?>", "")) \
    .drop("helpful", "summary", "review_text")

print("¡Pipeline ejecutado con éxito!")

# Mostrar el esquema resultante para verificar que esté en Primera Forma Normal (1FN)
print("\nEsquema limpio y normalizado en Spark:")
df_spark_clean.printSchema()

# Mostrar un "Sample" de los datos procesados al estilo Databricks
print("\nMuestra de los datos transformados:")
df_spark_clean.select("reviewer_id", "helpful_votes", "total_votes", "helpful_ratio", "word_count", "clean_summary").show(5, truncate=30)
## Trundado a 30 caracteres para evitar saturar la pantalla con textos largos en el resumen.
```

Para completar la infraestructura local del motor de Inteligencia Artificial antes de realizar el despliegue definitivo en los servidores cloud, seguiremos un orden estrictamente secuencial dividido en 4 fases de código:

[DF en Memoria Spark] ──> (1. Parquet) ──> (2. Chunking) ──> (3. Vector DB) ──> (4. LangChain RAG)

1. Persistencia Inmutable (Formato Parquet)
    * **Objetivo:**
        Guardar el DataFrame limpio de Spark en el disco duro local utilizando el formato de almacenamiento columnar **Parquet**.
    * **Propósito:**
        Experimentar cómo Big Data comprime la información, elimina la redundancia de esquemas del JSON y prepara los archivos para lecturas ultrarrápidas basadas en columnas específicas, emulando el almacenamiento en el Data Lake (o DBFS) de Databricks.

    * **Objetivo:** 
        Implementar un algoritmo de división de texto adaptativo para las reseñas limpias.
    * **Propósito:**
        Configurar un divisor de caracteres recursivo (`RecursiveCharacterTextSplitter`) parametrizado para crear fragmentos homogéneos (ej. 150 palabras con solape de 30). Esto evitará la pérdida de contexto en textos masivos y prevendrá el fenómeno de *"Lost in the Middle"* (pérdida de atención) en el modelo de lenguaje.

3. Gestión de la Base de Datos Vectorial (La Memoria de la IA)
    * **Objetivo:** 
        Desplegar una base de datos vectorial local (como **ChromaDB** o **FAISS**) orientada a embeddings de alta dimensionalidad.
    * **Propósito:** 
        Indexar los fragmentos de texto generados en el paso anterior y convertirlos en vectores matemáticos. Esta base de datos actuará como la memoria indexada donde el sistema realizará búsquedas de similitud del coseno.

4. Orquestación del Motor de IA con LangChain
    * **Objetivo:** 
        Conectar los componentes de extracción y generación (RAG - *Retrieval-Augmented Generation*).
    * **Propósito:** 
        Construir el pipeline final. Crearemos las plantillas de prompts de inteligencia de mercado, configuraremos el componente de recuperación (*Retriever*) conectado a la base vectorial y enviaremos el contexto curado al modelo de lenguaje (LLM) a través de Groq u Ollama para obtener las respuestas analíticas consolidadas.

```python
# Comparación de tamaños en disco entre el JSON original y la carpeta Parquet resultante
def get_dir_size(path):
    if os.path.isfile(path):
        return os.path.getsize(path)
    total_size = 0
    for dirpath, dirnames, filenames in os.walk(path):
        for f in filenames:
            fp = os.path.join(dirpath, f)
            if not os.path.islink(fp):
                total_size += os.path.getsize(fp)
    return total_size

size_json = get_dir_size("electronics_sample_30k.json") / (1024 * 1024)
size_parquet = get_dir_size("electronics_products_clean.parquet") / (1024 * 1024)

print("Análisis de Almacenamiento Local:")
print(f"Tamaño del JSON Crudo Original:  {size_json:.2f} MB")
print(f"Tamaño de la Carpeta Parquet:     {size_parquet:.2f} MB")
print(f"Reducción de espacio:            {((size_json - size_parquet) / size_json) * 100:.1f}% menos espacio en disco.")
```

### Segmentación Semántica

Cuando construimos un sistema RAG (Retrieval-Augmented Generation), el modelo de lenguaje (LLM) necesita consultar los datos para responder preguntas de mercado. Sin embargo, enviarle una reseña completa de cientos o miles de palabras a la base de datos vectorial tiene dos grandes problemas:

* **Ruido Semántico:**
    Una reseña larga puede hablar de la batería al inicio, de la pantalla en medio y del envío al final. Si la aglomeramos en un solo vector matemático, los conceptos se promedian y el buscador pierde precisión.

* **Límites de Contexto:**
    Los modelos de embeddings y los LLMs tienen un límite de tokens. Si les inyectamos documentos gigantescos, el costo de procesamiento se dispara y sufren del fenómeno "Lost in the Middle" (ignoran la información que está en el centro del texto).

Para resolver esto, aplicamos Chunking.

#### La Fórmula de Conversión Lingüística

En el idioma inglés una palabra promedio en textos de reseñas de internet tiene una longitud aproximada de 4 a 5 caracteres, a lo que debemos sumarle 1 carácter adicional por el espacio que la separa de la siguiente palabra.Por lo tanto, la métrica estándar de conversión se fija en:

$$\text{1 palabra} \approx 5.5 \text{ caracteres}$$

Si aplicamos esta constante a la configuración tenemos que el tamaño del fragmento (Chunk Size):

$$600 \text{ caracteres} \div 5.5 \approx 109 \text{ palabras}$$

*(Si el usuario escribe palabras cortas como "it", "good", "the", esos 600 caracteres se estiran fácilmente hasta alcanzar las 130 o 150 palabras reales).*

De forma análoga el traslape semántico (Chunk Overlap):

$$120 \text{ caracteres} \div 5.5 \approx 22 \text{ palabras}$$

Al configurar **chunk_size=600** y **chunk_overlap=120**, estamos cumpliendo con el objetivo de diseño: fragmentos lo suficientemente cortos para ser muy específicos, pero con el espacio necesario para que el modelo de embedding vectorice ideas completas.

```python
# 2. Proceso de segmentación semántica (Chunking)

# Leer el dataset limpio desde nuestra carpeta Parquet usando Spark

# Convertimos una muestra controlada a Pandas para la fase de diseño del Chunking
df_rag = df_spark_clean.select("asin", "reviewer_id", "clean_review_text").limit(1000).toPandas()

# Inicializar el divisor recursivo de LangChain
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=600,       # Tamaño máximo de caracteres por fragmento
    chunk_overlap=120,    # Traslape semántico entre fragmentos contiguos
    length_function=len,
    separators=["\n\n", "\n", " ", ""]
)

# Lista para almacenar los registros individuales segmentados
chunks_list = []

# Iterar sobre las reseñas para generar los fragmentos
for idx, row in df_rag.iterrows():
    review_text = row["clean_review_text"]
    
    # Validar que la reseña no esté vacía
    if review_text:
        # Despedazar el texto en fragmentos
        fragments = text_splitter.split_text(review_text)
        
        # Mapear cada fragmento manteniendo sus metadatos de origen (Trazabilidad)
        for i, fragment in enumerate(fragments):
            chunks_list.append({
                "asin": row["asin"],
                "reviewer_id": row["reviewer_id"],
                "chunk_id": f"{row['asin']}_{row['reviewer_id']}_chunk_{i}",
                "chunk_text": fragment
            })

# Consolidar en un nuevo DataFrame listo para IA
df_chunks = pd.DataFrame(chunks_list)

print("¡Segmentación completada con éxito!")
print(f"Reseñas originales procesadas:  {len(df_rag)}")
print(f"Fragmentos semánticos generados: {len(df_chunks)}")
print(f"Factor de expansión: {(len(df_chunks) / len(df_rag)):.2f} fragmentos por reseña en promedio.")

# Mostramos una muestra de cómo quedaron divididos los textos con su trazabilidad
print("\nInspección de fragmentos resultantes:")
print(df_chunks[["chunk_id", "chunk_text"]].head(3).to_string(max_colwidth=100))
```

### Gestión de la Base de Datos Vectorial

En un sistema RAG tradicional, el flujo convencional dictaría que tomemos cada texto, lo mandemos a una API de embeddings (como la de OpenAI) para convertirlo en un vector de alta dimensionalidad y luego guardemos esos vectores en la base de datos con extensión vectorial (como PostgreSQL).

Sin embargo, para la Fase 0 (Laboratorio Local), usar una base de datos pesada en un servidor externo o gastar dinero en APIs de embeddings para miles de fragmentos de prueba es ineficiente e innecesario.

Para resolver esto con costo cero, utilizaremos ChromaDB integrada con un modelo de embeddings que correrá de forma 100% local y gratuita dentro de la memoria RAM utilizando la librería sentence-transformers. Esto simulará el comportamiento de un clúster empresarial desde una computadora local.

Necesitaremos instalar las librerías `chromadb` y `sentence-transformers` necesarias en el entorno.

```python
# 3. Inicializando el motor de Embeddings Locales y ChromaDB

# Configurar el cliente de ChromaDB en Memoria RAM
chroma_client = chromadb.EphemeralClient()

# Definir el modelo de embeddings local usando Sentence Transformers (modelo ligero y eficiente para textos cortos)
sentence_transformer_ef = embedding_functions.SentenceTransformerEmbeddingFunction(
    model_name="all-MiniLM-L6-v2"
)   # Este modelo convierte cualquier texto en un vector matemático de 384 dimensiones

# Crear una Colección dentro de la base de datos vectorial
collection_name = "market_intelligence_reviews"
collection = chroma_client.create_collection(
    name=collection_name, 
    embedding_function=sentence_transformer_ef
)

print(f"Colección '{collection_name}' creada con éxito.")
```

```python
print("Ingestando y vectorizando fragmentos en la Base de Datos Semántica...")

# Preparar las listas que exige ChromaDB para una inserción masiva y eficiente
ids = df_chunks["chunk_id"].tolist()
documents = df_chunks["chunk_text"].tolist()

# Metadatos para que el buscador pueda filtrar por producto
metadata = df_chunks.apply(lambda r: {"asin": r["asin"], "reviewer_id": r["reviewer_id"]}, axis=1).tolist()

# Ingestar en ChromaDB (procesamos en un solo bloque ya que son 1,694 fragmentos)
collection.add(
    ids=ids,
    documents=documents,
    metadatas=metadata
)

print(f"\nSe han indexado y vectorizado {collection.count()} fragmentos en la memoria semántica de la IA.")
```

```python
# Simulación de una búsqueda de inteligencia de mercado

query_example1 = "charger stopped working and melted or got too hot"

print(f"Buscando en la memoria vectorial: '{query_example1}'...\n")

results = collection.query(
    query_texts=[query_example1],
    n_results=2 # Traer los 2 fragmentos más parecidos matemáticamente
)

# Desplegar los fragmentos más relevantes encontrados por cercanía semántica
for i, (doc, score, meta) in enumerate(zip(results['documents'][0], results['distances'][0], results['metadatas'][0])):
    print(f"Resultado #{i+1} (Distancia Matemática: {score:.4f})")
    print(f"ID: {meta['reviewer_id']} | Producto: {meta['asin']}")
    print(f"Texto extraído:\n{doc}\n")
    print("-" * 60)
```

#### Diagnóstico del Recuperador Semántico (ChromaDB)

Al ejecutar la consulta de prueba sobre componentes defectuosos o sobrecalentados, la base de datos vectorial devolvió fragmentos relacionados con cables dañados, pero con una **distancia matemática elevada (0.57 - 0.59)** y sin menciones explícitas a altas temperaturas. 

Este comportamiento no representa una falla del algoritmo, sino un comportamiento esperado bajo las siguientes condiciones de diseño:

* **Restricción de la Base de Conocimiento:** 
    Para esta fase de diseño, limitamos la ingesta a una muestra estricta de las primeras 1,000 reseñas (`.limit(1000)`). El sesgo estadístico de la muestra determinó la ausencia de quejas extremas por sobrecalentamiento en ese subconjunto.
* **Comportamiento del Umbral Semántico:** 
    En el modelo industrial `all-MiniLM-L6-v2`, distancias superiores a `0.55` indican que los vectores están alejados en el espacio hiperdimensional. El motor cumplió con su directiva de extraer lo "más cercano disponible" (conceptos de cables/cargadores con problemas de uso), demostrando la efectividad de la búsqueda por similitud del coseno.

### Orquestación del Motor de IA con LangChain

Llegamos al último eslabón de la cadena local. Vamos a construir el circuito de Generación Aumentada por Recuperación (RAG).

El flujo es el siguiente

1. El usuario hace una pregunta analítica.
2. El sistema consulta la DB y extrae el contexto.
3. Inyectamos ese contexto dentro de un Prompt de Ingeniería de Mercado.
4. Le pasamos todo empaquetado al LLM para que actúe como un consultor experto y redacte la conclusión.

Para mantener el laboratorio 100% libre, local y sin depender de una API Key en esta fase de prueba encapsularemos la llamada para ver cómo se estructura el prompt final y cómo se le presentaría al LLM en producción.

```python
# 4. Configurando el Orquestador RAG con LangChain

# Definir la consulta de negocio del usuario
user_question = "What are the main quality complaints regarding charging cables and power bricks?"

# Extraer el contexto de la Base Vectorial (Retriever)
result_vectors = collection.query(
    query_texts=[user_question],
    n_results=3
)
extracted_context = "\n\n".join(result_vectors['documents'][0])

# Diseño del Prompt de Ingeniería de Mercado (System Prompt)
system_prompt = f"""
You are an expert Market Intelligence AI Assistant. Your task is to analyze customer reviews 
and synthesize a professional summary regarding product defects, quality issues, or patterns of failure.

Use ONLY the following pieces of retrieved context from customer reviews to answer the user's question. 
If the context does not contain enough information, state that clearly.

---
RETIREVED CUSTOMER REVIEWS (CONTEXT):
{extracted_context}
---

USER QUESTION:
{user_question}

INSTRUCTIONS:
1. Be concise, technical, and objective.
2. Focus strictly on quality flaws or product behavior mentioned in the context.
3. Do not make up facts or extrapolate beyond the provided text.

EXPERT ANALYSIS:
"""

print("Pipeline RAG ensamblado de forma correcta.")
print("\nPrompt final que se le enviará al LLM en Databricks:\n")
print(system_prompt)
```
